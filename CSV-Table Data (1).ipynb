```python
import os
import pandas as pd

csv_path = "weekend_decisions.csv"

PURPLE_BG = "\033[48;2;75;0;130m"
WHITE_FG = "\033[38;2;255;255;255m"
DARK_GRAY_BG = "\033[48;2;169;169;169m"
MID_GRAY_BG = "\033[48;2;211;211;211m"
TEXT_FG = "\033[38;2;0;0;0m"
RESET = "\033[0m"

if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    col_widths = [
        max(df[col].astype(str).map(len).max(), len(col)) + 4
        for col in df.columns
    ]

    print("\n")
    header_str = ""
    for i, col in enumerate(df.columns):
        header_str += f"{col.center(col_widths[i])}"
    print(f"{PURPLE_BG}{WHITE_FG}{header_str}{RESET}")

    for row_idx, row in df.iterrows():
        row_str = ""
        for col_idx, value in enumerate(row):
            row_str += f"{str(value).center(col_widths[col_idx])}"
        current_bg = MID_GRAY_BG if row_idx % 2 == 0 else DARK_GRAY_BG
        print(f"{current_bg}{TEXT_FG}{row_str}{RESET}")
    print("\n")
else:
    print(f"Error: {csv_path} not found.")
```
